```python
from pathlib import Path
import pandas as pd
```

```python
ROOT_DIR = Path("../")
DATA_DIR = ROOT_DIR / "data"
RAW_DATA_DIR = DATA_DIR / "00_raw" / "text"
OUTPUT_DATA_ROOT_DIR = DATA_DIR / "01_processed" / "Suryaghar Call Analysis"
OUTPUT_DATA_ROOT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
df_1 = pd.read_csv(RAW_DATA_DIR / "pm_suryaghar_stage_wise_report_20260423_152452.csv")
df_2 = pd.read_csv(RAW_DATA_DIR / "pm_suryaghar_stage_wise_report_20260423_152457.csv")
```

```python
df = pd.concat([df_1, df_2], ignore_index=True)
```

```python
df
```

```python
df["Remark"].value_counts()
```

```python
# drop all duplicates in the Remark column from df
df_unique_remark = df.drop_duplicates(subset=["Remark"])
```

```python
df
```

```python
# get one long string from Remarks with each remark separated by incrementing number and on new line
df_unique_remark["numbered_remarks"] = df_unique_remark.index.astype(str) + ": " + df_unique_remark["Remark"]
```

```python
txt = df_unique_remark["numbered_remarks"].str.cat(sep="\n")
with open(OUTPUT_DATA_ROOT_DIR / "numbered_remarks.txt", "w") as f:
    f.write(txt)
```
